## 1. Đếm xem có bao nhiêu dòng và bao nhiêu dòng xuất hiện spam và bao nhiêu dòng xuất hiện ham trong file trên

```python
# Đọc dữ liệu
with open("SMSSpamCollection", "r", encoding="utf8") as f:
    lines = f.readlines()

# Tổng số dòng
total_lines = len(lines)

# Đếm spam và ham
spam_count = 0
ham_count = 0

for line in lines:
    if line.startswith("spam"):
        spam_count += 1
    elif line.startswith("ham"):
        ham_count += 1

print("Tổng số dòng:", total_lines)
print("Số dòng spam:", spam_count)
print("Số dòng ham:", ham_count)
```

## 2. Đếm xem có bao nhiêu từ (các từ ngăn cách bởi khoảng trắng, và không phải chữ số)

```python
import re

# Ghép toàn bộ nội dung thành 1 chuỗi
text = " ".join(lines)

# Lấy các từ KHÔNG phải chữ số
words = re.findall(r"[A-Za-z]+", text)

print("Tổng số từ:", len(words))
```


## 3. Đếm xem có bao nhiêu từ 'you' (không phân biệt hoa thường)

```python
count_you = sum(1 for w in words if w.lower() == "you")
print("Số lần xuất hiện từ 'you':", count_you)
```


## 4. Đếm xem có bao nhiêu từ khác nhau (không phân biết hoa thường)

```python
unique_words = set(w.lower() for w in words)
print("Số từ khác nhau:", len(unique_words))
```


## 5. Liệt kê 10 từ khóa xuất hiện nhiều nhất trong file dữ liệu (không phân biết hoa thường)

```python
from collections import Counter

# Đếm tần suất
freq = Counter(w.lower() for w in words)

# Lấy 10 từ xuất hiện nhiều nhất
top10 = freq.most_common(10)

print("10 từ xuất hiện nhiều nhất:")
for word, count in top10:
    print(f"{word}: {count}")
```
