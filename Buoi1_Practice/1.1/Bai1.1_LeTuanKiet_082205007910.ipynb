### 3.3 Set

**M√¥ t·∫£:** Set l√† t·∫≠p h·ª£p kh√¥ng c√≥ th·ª© t·ª±, kh√¥ng ch·ª©a gi√° tr·ªã tr√πng, v√† c√≥ th·ªÉ thay ƒë·ªïi.

**Ng·ªØ c·∫£nh √°p d·ª•ng:**
- Lo·∫°i b·ªè ph·∫ßn t·ª≠ tr√πng l·∫∑p
- Th·ª±c hi·ªán c√°c ph√©p to√°n t·∫≠p h·ª£p (union, intersection, difference)
- Ki·ªÉm tra th√†nh vi√™n nhanh (to√°n t·ª≠ in c·ª±c nhanh tr√™n set)

**V√≠ d·ª• minh h·ªça:**
